```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
# Generate synthetic dataset
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
```

```python
# Initialize KMeans with the desired number of clusters (e.g., 4)
kmeans = KMeans(n_clusters=4, random_state=0)
```

```python
# Fit the model to the data
kmeans.fit(X)
```

```python
# Predict the cluster for each data point
y_kmeans = kmeans.predict(X)
```

```python
# Plot the data points with their respective cluster color
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
```

```python
# Plot the centroids of the clusters
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')

# Display the plot
plt.title('K-means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate synthetic dataset
n_samples = 300
n_features = 2
n_clusters = 3
random_state = 42

# Create dataset with blobs
points, true_labels = make_blobs(n_samples=n_samples, n_features=n_features, centers=n_clusters, random_state=random_state)

# KMeans clustering using sklearn
kmeans = KMeans(n_clusters=n_clusters, random_state=random_state)
kmeans.fit(points)

# Get the cluster centers and labels
cluster_centers = kmeans.cluster_centers_
predicted_labels = kmeans.labels_

# Plotting the results
plt.figure(figsize=(8, 6))

# Plot the points with colors based on their predicted cluster
plt.scatter(points[:, 0], points[:, 1], c=predicted_labels, cmap='viridis', marker='o', s=50)

# Mark the cluster centers
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], c='red', marker='X', s=200, label='Centroids')

plt.title('KMeans Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()
```
